```python
categorical_columns = [
    "History of HeartDisease or Attack", "Told High Cholesterol", "Smoked 100+ Cigarettes",
    "Diagnosed Diabetes", "Doctor Visit Cost Barrier", "General Health", "Difficulty Walking"
]
df[categorical_columns] = df[categorical_columns].fillna("Unknown")
```

```python
binary_columns = [
    "History of HeartDisease or Attack", "High Blood Pressure", "Told High Cholesterol",
    "Cholesterol Checked", "Smoked 100+ Cigarettes", "Diagnosed Stroke", "Diagnosed Diabetes",
    "Leisure Physical Activity", "Heavy Alcohol Consumption", "Health Care Coverage",
    "Doctor Visit Cost Barrier", "Difficulty Walking", "Vegetable or Fruit Intake (1+ per Day)"
]
```

```python
df[binary_columns] = df[binary_columns].apply(lambda x: x.map({'Yes': 1, 'No': 0, 'Unknown': -1}))
df
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df["Body Mass Index"], bins=30, kde=True, color='blue')
plt.xlabel("Body Mass Index")
plt.ylabel("Count")
plt.title("Distribution of Body Mass Index with KDE")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["Body Mass Index"], color='blue')
plt.xlabel("Body Mass Index")
plt.title("Boxplot of Body Mass Index")
plt.show()
```

```python
df['Body Mass Index'] = df['Body Mass Index'].fillna(df['Body Mass Index'].median()) # Careful for Personal/Sensitive Data
```

```python
df = df.dropna(subset = 'Body Mass Index')
```

```python
# Outliers range
Q1 = df["Body Mass Index"].quantile(0.25)
Q3 = df["Body Mass Index"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df_outliersdetected = df[(df["Body Mass Index"] >= lower_bound) & (df["Body Mass Index"] <= upper_bound)]
```

```python
df['Body Mass Index'].describe()
```

```python
df = df[(df["Body Mass Index"] >= lower_bound) & (df["Body Mass Index"] <= 60)]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_outliersdetected["Body Mass Index"], color='blue')
plt.xlabel("Body Mass Index")
plt.title("Boxplot of Body Mass Index")
plt.show()
```

```python
df = df.reset_index(drop = True)
```

```python
education_mapping = {
    "Never attended school" : 1,
    "Elementary" : 2,
    "Some high school": 3,
    "High school graduate": 4,
    "Some college or technical school": 5,
    "College graduate": 6
}
df['Education Level'] = df['Education Level'].map(education_mapping)

# Encoding General Health (Ordinal)
health_mapping = {
    "Very Poor": 1,
    "Poor": 2,
    "Fair": 3,
    "Good": 4,
    "Excellent" : 5
}
df['General Health'] = df['General Health'].map(health_mapping)
```

```python
le = LabelEncoder() #low, high, peak -> 1, 2, 3
df['Sex'] = le.fit_transform(df['Sex'])
```
