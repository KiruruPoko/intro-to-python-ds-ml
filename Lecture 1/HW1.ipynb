$$
A = \begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
7 & 8 & 9
\end{bmatrix}
\qquad
B = \begin{bmatrix}
7 & 9 & 3 \\
3 & 3 & 1 \\
2 & 1 & 4
\end{bmatrix}
$$


การคูณเมทริกซ์ด้วยเมทริกซ์ แล้วทรานสโพส (Transpose) มีสมบัติดังนี้

$$
(AB)^T = A^T B^T
$$
