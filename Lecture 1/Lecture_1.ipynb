```python
# Dictionaries
d = {1: 'Geeks', 2: 'For', 3: 'Geeks', 'age':22}

# Using del to remove an item
del d["age"]
print(d)

# Using pop() to remove an item and return the value
val = d.pop(1)
print(val)

# Using popitem to removes and returns
# the last key-value pair.
key, val = d.popitem()
print(f"Key: {key}, Value: {val}")

# Clear all items from the dictionary
d.clear()
print(d)
```

c^2 = a^2 + b^2

```python
# Function
def pythagorus(a,b):
  return (a**2 + b**2) ** (1/2)
print(pythagorus(3, 4))
```

```python
# Lambda function
pythagorus = lambda a, b: (a**2 + b**2) ** (1/2)
print(pythagorus(3, 4))
```

## Basic Operations

Arithmetic operators on arrays apply elementwise. A new array is created and filled with the result.

```python
a = np.array([2, 40, 73, 87])

b = np.array([1, 20, 20, 50])

c = a - b

print(c)
```

```python
d = a ** 2

print(d)
```

```python
e = np.array([0, np.pi, 2*np.pi])

f = np.cos(e)

print(f)
```

```python
g = np.array([10, 20 ,30 ,40])

h = g < 30

print(h)
```

Unlike in many matrix languages, the product operator * operates elementwise in NumPy arrays. The matrix product can be performed using the @ operator (in python >=3.5) or the dot function or method:



```python
A = np.array([[1, 1], [0, 1]])

B = np.array([[2, 0], [3, 4]])

elementwise_product = A*B # elementwise product

print(elementwise_product)
```

Let's try this!

http://matrixmultiplication.xyz/

```python
matrix_product = A@B # matrix product

print(matrix_product)
```

```python
another_matrix_product = A.dot(B) # another matrix product

print(another_matrix_product)
```

## Indexing, Slicing and Iterating

One-dimensional arrays can be indexed, sliced and iterated over, much like lists and other Python sequences.

```python
a = np.arange(10)**3

print(a)
```

```python
a[2]
```

```python
a[2:5]
```

```python
a[ : :-1]
```

Multi dimentional array.

```python
A = np.array([[ 0,  1,  2,  3],
       [10, 11, 12, 13],
       [20, 21, 22, 23],
       [30, 31, 32, 33],
       [40, 41, 42, 43]])

print(A)
```

```python
A[2,3]
```

```python
A[0:5, 1] # each row in the second column of A
```

```python
A[ : ,1] # equivalent to the previous example
```

```python
A[1:3, : ] # each column in the second and third row of b
```

```python
A[-1] # the last row. Equivalent to A[-1,:]
```
